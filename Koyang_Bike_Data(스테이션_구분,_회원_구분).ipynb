```python
# 중복 데이터를 제거하기 위해 함수 사용 (중복되지 않는 데이터를 나타낸다.)
def remove_duplicated(a, b):                                                    
    c = set(a) | set(b)                                                         
    d = set(a) & set(b)                                                         
    e = c - d                                                                   
    return e     
```

```python
a = remove_duplicated(df_bike_sta, df_rental_sta) #17-19 데이터에는 사용 현황이 없음 / 998 - 현재 자전거 스테이션에는 없음
b = remove_duplicated(df_bike_sta, df_return_sta) #17-19 데이터에는 사용 현황이 없음 / 0, 998, 999 - 현재 자전거 스테이션에는 없음. 
c = remove_duplicated(df_return_sta, df_rental_sta)
print('현재 스테이션 - 대여 스테이션:', a) 
print('현재 스테이션 - 반납 스테이션:', b)
# df_Bike 에서 스테이션 번호 확인해보기
print('대여 - 반납 스테이션:', c) # 0,999 어떤 것들이 있는지 확인
```
